10 - Peça a quantidade de horas trabalhadas e o valor por hora. Se trabalhou até 40 horas, calcule o salário normalmente (horas × valor por hora). Se trabalhou mais de 40 horas, as horas extras devem ter um adicional de 50% (horas extras × valor por hora × 1.5). Crie um algoritmo em Python que realize esse cálculo corretamente.

```python
horas_trabalhadas = float(input("Digite a quantidade de horas trabalhadas: "))
valor_hora = float(input("Digite o valor recebido por hora: "))

if horas_trabalhadas <= 40:
    salario = horas_trabalhadas * valor_hora
elif horas_trabalhadas > 40:
    horas_extras = horas_trabalhadas - 40
    salario = (40 * valor_hora) + (horas_extras * valor_hora * 1.5)
else:
    salario = 0

print(f"O salário total é R$ {salario:.2f}")
```

##Jogo da Adivinhação

```python
import random as rd
```

```python
aleatorio = rd.randint(0,100)

print("============================================")
print("            JOGO DE ADIVINHAÇÃO             ")
print("============================================")
print("Tente adivinhar o número que estou pensando!")
print("============================================")

tentativa = 5

while(tentativa >0):
  numero = int(input("Dígite um número: "))
  if (numero == aleatorio):
    print("Parábens! Você acertou!")
    break
  elif (numero > aleatorio):
    print("Digíte um numero menor: ")
  else:
    print("Digíte um número maior: ")
  tentativa -= 1
if (tentativa == 0):
  print("Você perdeu!")
```
