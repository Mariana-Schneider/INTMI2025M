1 - Crie um algoritimo que receba 2 números, em váriaves diferentes (num1 - num2), depois faça as quatro operações com esses 2 números e mostre o resultado na tela.

```python
num1 = int(input("Digite um número: ") )
num2 = int(input("Digite um número: "))
print("A soma é: ", num1+num2)
print("A subtração é: ", num1-num2)
print("A multiplicação é: ", num1*num2)
print("A divisão é: ", num1/num2)
```

```python
print(f"O primeiro número é {num1} e o segundo é {num2}")
```

3 - Crie um algoritimo que receba o nome do aluno, e 3 notas. Depois calcule a média das notas e imprima na tela:
O aluno (aluno) etá com (média) de média!

```python
aluno = str(input("Digite seu nome: "))
n1 = float(input("1º nota: "))
n2 = float(input("2º nota: "))
n3 = float(input("3º nota: "))
média = n1+n2+n3/3
print(f"O aluno {aluno} etá com {média: .2f} de média!")
```

1 - Peça um número e informe se ele é maior que 10. Crie um algoritmo em Python para fazer essa verificação.

```python
t = int(input("Digite um número: "))
if t >= 10:
    print(f"{t} é maior ou igual a 10")
else:
    print(f"{t} é menor que 10")
```

2 - Peça um número inteiro e informe se ele é par. Crie um algoritmo em Python que verifique se o número é divisível por 2.

```python
u = int(input("Digite um número: "))
if u % 2 == 0:
    print(f"{u} é um número par")
else:
    print(f"{u} não é um número par")
```
